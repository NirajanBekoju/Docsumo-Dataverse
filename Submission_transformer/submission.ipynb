# Load the label encoder

we need label encoder to transform the test data category into label and verify the classification report

```python
label_encoder_path = folder_path + 'label_encoder.pickle'
with open(label_encoder_path, 'rb') as handle:
    label_encoder = pickle.load(handle)
```

# Load the preprocessed test data

# load the preprocessed test data obtained above.

```python
test_file_path = "/content/drive/MyDrive/Colab Notebooks/DataVerse/preprocessed_test_version_2_lemmatizer.csv" 
test_df = pd.read_csv(test_file_path)
```

```python
test_df.info()
```

tokenize the test data and padd the tokenized data

```python
X_test = test_df.text
test_sequences = tokenizer.texts_to_sequences(X_test)
print(test_sequences[0])
test_padded = pad_sequences(test_sequences, maxlen = MAX_PAD_LENGTH, padding = 'post', truncating = 'post')
print(test_padded[0])
```

use the loaded model to predict the data

```python
y_test_predict = loaded_model.predict(test_padded)
y_predicted = np.argmax(y_test_predict, axis = 1)
```

obtain the true label by label encoding the true test category

```python
y_true = label_encoder.transform(test_df.category)
```

Obtain the classification report

```python
from sklearn.metrics import classification_report, f1_score
```

```python
f1_score(y_true, y_predicted, average = "macro")
```

```python
print(classification_report(y_true, y_predicted))
```
